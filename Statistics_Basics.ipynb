Q1.  Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales

A1. There are 2 types of data:

1. Qualitative Data or Categorical Data: These kinds of data are non numeric in nature and can be further classified into 2 subtypes.

a. Nominal data: These are data without any order, for example blood group, gender race ethinicity.

b. Ordinal data: These kind of qualitative data depict an order among the given feature in the data frame. For example Rank, grades, etc.

2. Quantitative data: These are data that are numerical in nature and hence qualtifiable. They can be further subdivided into.

a. Discrete: These are essentially represented by while numbers. For example. number of children, number of rooms etc

b. Continuous: These are represented by real numbers. for example height, weight, distance etc.

Corresponding to these types of data there are their respective measuring scales:

1. Nominal scale data: These are the qualitative or categorical scaled data. They have no order among them and have no true '0' starting point. they can be represented through a pie chart and bar graph

2. Ordinal scale data: In these kind of data the rank or the order has meaning but their difference cannot be measured and it does not have a true starting  point

3. Intervam measure scale: These kind of data are labelled, has a meaningful order and their difference can be measured. However, they do not have a zero starting point, therefore their difference cannot be measured in case of ratios

4. Ratio scaled data: In these kind of data, the order and rank has a meaning and there difference can be measured, also they have a true zero starting point.

Q2. What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.

A2.Central tendency is defined as the representation of the central point of a set of data values. The most frequently used measures of central tendencies are mean median and mode.

Mean: It is the arithmetic average of the given set of data. It is more accurate for a numerical set of data where there are no outliers.

Median: It is the physical mean of the given set of data. In this scenario, the given set of data is arranged in ascending order and then the mid value of the set is said to be median. It is used more often in case of a set of data where there are outliers.

Mode: Mode is refered to the item, in a given set of data, which has the highest frequency. Mode is usualy used in place of a categorical data.

```python
#mean median mode
import numpy as np
data=[21,32,43,21,54,65,76,87,8000]
print("Mean=",np.mean(data))
print("Median=",np.median(data))
import statistics
print("Mode=",statistics.mode(data))
```

Q3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data.

A3. Dispersion refers to the extent to which the items in a given set of data deviate from the mean.

Variance is one of the ways in which the measure of dispersion is achieved. It is the average of the aquared difference from the mean and measures the spread at an overall level.

Standard deviation is mathematically represented as the square root of variance and represents the deviation of the items from their mean at an individual level.


```python
data=[21,32,43,21,54,65,76,87]
np.var(data)
```

```python
np.std(data)
```

Q6.  Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

A6. Skewness is the measure of symmetricity of data. The level of assymetricity is called skewness. Skewness might also represent that outliers are present within the data set.

Not skewed: It means that there are equal number of data on both sides of the mean. It means mean, median and mode are equal.

Positively skewed: It means that there are more number of data points on one side of the mean than on the other. In other words that more number of data points are greater than the mode than those are lesser than it.

Negatively skewed: It means that there are more number of data points on one side of the mean than on the other. In other words that more number of data points are lesser than the mode than those are greater than it.

```python
data=[21,32,43,54,54,54,65,76,87] #skewness=0
import numpy as np
import seaborn
sns.displot(data, kde=True)
```

Q7.  What is the interquartile range (IQR), and how is it used to detect outliers?

A7. Outliers are extreme values which affect the mean and the range of the given data set. Therefore to negate the effect of the outliers on the mean and the overall range of the given data set the concept of Interquartile range (IQR) arises.

IQR is the difference between the 75th percentile and the 25th percentile value of the given data set. It helps one to get a range of the data without considering the outliers.

Q1= 25 percentile
Q2= 50 percentile
Q3= 75 percentile

IQR=Q3-Q1

Upper Whisker= Q3+1.5*IQR

Lower Whisker= Q1-1.5*IQR

Any data point that is above or below the upper and the lower whisker respectively is treated as an outlier.

Q9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).

A9. The key features of normal distribution are as follows:

1. Symmetricity: The normal distribution should be symmetric on both sides. That means that 50% of the data points should be greater than that of the mean and 50% of the data points should be lesser than that of the mean.

2. Bell shaped curve: The normal distribution is a bell shaped curve where the cluster around the peak represents most of the data points.

3. Mean Median and mode in a normal distribution represent the same point or value.

4. Standard deviation: The spread of the distribution is determined by standard deviation. The first deviation is represented by 68 % of the data points. 95% of the data are represented within 2 standard deviations and 99.7% of the data are represented by 3 standard deviations.
